```python
import GetOldTweets3 as got
import datetime
import pandas as pd
import urllib.request as urllib2
from urllib.error import HTTPError
import time
import pickle
import retrying
import random
import os


full=pd.read_csv("county_centers.csv")
center=pd.DataFrame()
center['counties']=full['County [2]']
center['long']=full['Longitude']
center['lat']=full['Latitude']
start_date = datetime.date(2019, 12, 31)

end_date = datetime.date(2020, 6,1)

delta = datetime.timedelta(days=1)
i=0
inf_count=0

n_tweets_per_day = 100
target_folder = "/Volumes/Extreme/100"

@retrying.retry(wait_fixed=10000)       
def retry_got(tweetCriteria):
    #print("trying")
    tweets = got.manager.TweetManager.getTweets(tweetCriteria)
    return tweets
    

while inf_count<=10000:
    os.chdir(target_folder)
    files_no_ext = [".".join(f.split(".")[:-1]) for f in os.listdir() if os.path.isfile(f)]
    os.chdir("..")
    
    
    i=random.randint(0, len(center.counties))
    if str(i) in files_no_ext:
        print("duplicate")
        continue
    print("Fetching tweets for county {}".format(i))
        
        
    df_tweets=pd.DataFrame(columns={str(i)})
    tweet_dict = {}

            
    while start_date <= end_date:
        print("Date: {}".format(start_date))
        start_date += delta
        df_tweets=df_tweets.append(pd.Series(name=start_date))
  
    
        location=str(center.lat[i])+", "+str(center.long[i])
        location = location.replace("°", "")
        location = location.replace("+", "")
        location=location.replace("\u2013", "-") #en dash

        #print(str(start_date))
        tweetCriteria = got.manager.TweetCriteria().setQuerySearch('*')\
                                           .setNear(location)\
                                           .setUntil(str(start_date))\
                                           .setMaxTweets(n_tweets_per_day)\
                                           .setWithin("40km")
                                                                      

        tweets=retry_got(tweetCriteria)
    
  
        for tweet in tweets:
            tweet_dict.setdefault(start_date, []).append(tweet)
    
    # save
    pickle.dump(tweet_dict, open(target_folder + '/' + str(i) + '.p', "wb"))
    start_date = datetime.date(2019, 12, 31)

    end_date = datetime.date(2020, 6, 1)
    inf_count=inf_count+1
```
